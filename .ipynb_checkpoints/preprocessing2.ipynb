## Task 2: Cleaning the Data

-  ### Subtask 2.1: Inspection and filtering movies data in 2017 

```python
pd.to_datetime(meta_data['release_date'], errors='coerce')
```

```python
meta_data['release_date'] = pd.to_datetime(meta_data['release_date'],format="%Y-%m-%d",errors='coerce')
```

```python
meta_data['year'] = pd.DatetimeIndex(meta_data['release_date']).year
```

```python
meta_data['year'].value_counts().sort_index(ascending=False)
```

```python
# Getting 2017 movies
new_data = meta_data.loc[meta_data.year==2017,['id','title','genres']]
```

```python
new_data
```

```python
# replacing type of `id` columns in `new_data` and `credits_data`
new_data['id'] = new_data['id'].astype(int)
```

```python
credits_data['id'] =  credits_data['id'].astype(int,errors='ignore')
```

- ### Subtask 2.2: mergence movies data of 2017 and creadits 

```python
data = pd.merge(new_data, credits_data, on='id')
```

```python
data.head(5)
```

```python
data['genres'][0]
```

```python
# convert s string to Python list
data['genres'] = data['genres'].apply(literal_eval)
data['cast']=data['cast'].apply(literal_eval)
data['crew'] =data['crew'].apply(literal_eval)
```

```python
data['genres'][0]
```

```python
# function get name of genres to string
def get_genres(x):
  list=[]
  st=" "
  for i in x:
    if i['name']=='Science Fiction':
      scifi = 'Sci-Fi'
      list.append(scifi)
    else :
      list.append(i['name'])
  if list ==[]:
    return np.NaN
  else :
    return st.join(list)
```

```python
data['genres']=data['genres'].map(lambda x :get_genres(x))
```

```python
# function get name of actors to string
def get_actor(x):
  list=[]
  str ='|'
  for i in x:
      list.append(i['name'])
  if list ==[] :
    return np.NaN
  else:
    return str.join(list[:3])
```

```python
 casts = data['cast'].apply(lambda x: get_actor(x))
```

```python
casts
```

```python
 data[['actor_1_name', 'actor_2_name', 'actor_3_name']] = casts.str.split('|', expand = True)
```

```python
data[['actor_1_name', 'actor_2_name', 'actor_3_name']]
```

```python
#function direction name to string
def get_directors(x):
  list=[]
  str =" "
  for i in x:
    if i['job']=='Director':
      list.append(i['name'])
  if list==[]:
    return np.NaN
  else:
    return str.join(list)
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

```python
data.head(10)
```

```python
data.columns
```

```python
#filtering features
data = data.loc[:,['title','director_name', 'genres','actor_1_name', 'actor_2_name',
       'actor_3_name']]
```

-  ### Subtask 2.3: Dropping unecessary rows

Some of the rows might have greater than five NaN values, should be removed.

```python
# checking null values
data.isnull().sum(axis=1).sort_values(ascending=False)
```

```python
# inspection the rows here if null columns 
(data.isnull().sum(axis=1).sort_values(ascending=False)).sum()
```

```python
round(data.isnull().sum()/len(data)*100,2)
```

```python
# dropping null values
data = data.dropna(how='any')
```

```python
data = data.rename(columns={'title':'movie_title'})
```
