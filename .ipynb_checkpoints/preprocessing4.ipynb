```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '1974662aed9fdffb3ac0d19b14b68349'
from tmdbv3api import Movie
tmdb_movie = Movie()
```

```python
Movie().search('The Grudge')
```

```python
#function get genre from api tmdb
def get_genre(x):
  genres = []
  result = tmdb_movie.search(x)
  if not result:
      return np.NaN
  else:
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
      genres_str = " "
      for i in range(0, len(data_json['genres'])):
          genres.append(data_json['genres'][i]['name'])
      return genres_str.join(genres)
    else:
      return np.NaN
```

```python
get_genre('The Grudge')
```

```python
data_20['genres'] = data_20['Title'].map(lambda x: get_genre(str(x)))
```

```python
df_20 = data_20[['Title','Cast and crew','genres']]
```

```python
df_20['Cast and crew'][0]
```

```python
# function get director
def get_director(x):
  if " (director)" in x:
    return x.split(" (director)")[0]
  elif " (directors)" in x:
    return x.split(" (directors)")[0]
  else:
    return x.split(" (director/screenplay)")[0]
  
```

```python
director = df_20['Cast and crew'].map(lambda x: get_director(x))
```

```python
df_20['director_name'] = director
```

```python
def get_actor(x):
  return ((x.split("screenplay); ")[-1]).split(", "))[:3]
```
